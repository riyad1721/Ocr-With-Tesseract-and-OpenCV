```python
import cv2
from matplotlib import pyplot as plt
import pytesseract
```

```python
def plot_two_images(img1, img2, title1="", title2=""):
    fig = plt.figure(figsize=[15,15])
    ax1= fig.add_subplot(121)
    ax1.imshow(img1, cmap="gray")
    ax1.set(xticks=[], yticks=[], title=title1)
    ax2= fig.add_subplot(122)
    ax2.imshow(img2, cmap="gray")
    ax2.set(xticks=[], yticks=[], title=title2)
```

```python
image = cv2.imread('data/image-4.jpeg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
thresh = 255 - cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)[1]
thresh = cv2.GaussianBlur(thresh, (5,5), 0)
plot_two_images(image, thresh, "real image", "thresh image")
```

# Only English document


```python
data = pytesseract.image_to_string(thresh, lang='eng', config='--psm 6')
print(data)
```

# Only Bangla document

```python
data = pytesseract.image_to_string(thresh, lang='ben', config='--psm 6')
print(data)
```

```python
data = pytesseract.image_to_string(thresh, lang='eng+ben', config='--psm 6')
print(data)
```
